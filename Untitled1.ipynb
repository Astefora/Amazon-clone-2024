```python
# 1. Upload CSV from your computer
from google.colab import files
uploaded = files.upload()  # Upload rivers.csv

# 2. Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             classification_report, confusion_matrix,
                             roc_auc_score, roc_curve)
import matplotlib.pyplot as plt
import seaborn as sns

# 3. Load the uploaded CSV
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name, encoding='latin1')
print("Initial Data Shape:", data.shape)
display(data.head())

# 4. Drop missing values
data = data.dropna()

# 5. Clean 'M[E] (metric tons year -1)' column
data['M[E] (metric tons year -1)'] = (
    data['M[E] (metric tons year -1)']
    .astype(str)
    .str.replace("'", "")
    .str.replace(",", "")
    .str.replace("%", "")
)
data['M[E] (metric tons year -1)'] = pd.to_numeric(data['M[E] (metric tons year -1)'], errors='coerce')
data = data.dropna(subset=['M[E] (metric tons year -1)'])

# 6. Create binary target column
data['plastic_contribution'] = data['M[E] (metric tons year -1)'].apply(lambda x: 0 if x > 6008 else 1)

# 7. Plot number of rivers by plastic contribution class
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='plastic_contribution', palette='Set2')
plt.xticks([0, 1], ['High Contribution (0)', 'Low Contribution (1)'])
plt.title('Plastic Contribution Class vs Number of Rivers')
plt.xlabel('Plastic Contribution Class')
plt.ylabel('Number of Rivers')
plt.tight_layout()
plt.show()

# 8. Clean all numeric columns
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].str.replace("'", "").str.replace(",", "").str.replace("%", "")
        data[col] = pd.to_numeric(data[col], errors='coerce')

# 9. Prepare features (X) and target (y)
X = data.drop(columns=[
    'M[E] (metric tons year -1)', 'plastic_contribution',
    'River name', 'Country or Administrative area'
], errors='ignore')
data = data.dropna(subset=X.columns)
X = data[X.columns]
y = data['plastic_contribution']

# 10. Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 11. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

# 12. Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 13. Predict
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]  # for ROC curve

# 14. Evaluation metrics
print("\nModel Evaluation:")
print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 15. Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# 16. ROC/AUC curve
roc_auc = roc_auc_score(y_test, y_proba)
fpr, tpr, _ = roc_curve(y_test, y_proba)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, color='darkorange', label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()

# 17. Feature importance
importances = model.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

print("\nFeature Importances:")
for i in indices:
    print(f"{features[i]}: {importances[i]:.4f}")

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
sns.barplot(x=importances[indices], y=[features[i] for i in indices], palette="viridis")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
